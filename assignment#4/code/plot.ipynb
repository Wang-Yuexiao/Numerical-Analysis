```python
import numpy as np
import matplotlib.pyplot as plt

def read_data(filename):
    x = []  
    y = []  
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            x.append(float(parts[0]))  
            y.append(int(parts[1]))    
    return x, y

def plot_histogram(x, y, title):
    plt.figure(figsize=(10, 6))
    plt.bar(x, y, width=(x[1] - x[0]), edgecolor='black')
    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()

def simulate_data(filename, distribution_type, sample_size, a=-3, b=4, m=0.5, s=1.5):
    bins = np.linspace(a, b, 100)  
    data = []
    
    if distribution_type == "uniform":
        data = np.random.uniform(a, b, sample_size)
    elif distribution_type == "gaussian":
        data = np.random.normal(m, s, sample_size)
    
    hist, edges = np.histogram(data, bins=bins)
    
    with open(filename, 'w') as f:
        for i in range(len(hist)):
            f.write(f"{edges[i]:.6f} {hist[i]}\n")

simulate_data("uniform_100.txt", "uniform", 100)
simulate_data("gaussian_100.txt", "gaussian", 100)

x_uniform, y_uniform = read_data("uniform_100000.txt")
plot_histogram(x_uniform, y_uniform, "Uniform Distribution Histogram")

x_gaussian, y_gaussian = read_data("gaussian_100000.txt")
plot_histogram(x_gaussian, y_gaussian, "Gaussian Distribution Histogram")
```
